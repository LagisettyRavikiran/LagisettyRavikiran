```python
#importing libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers,models
import matplotlib.pyplot as plt
import cv2
import os
import PIL
import numpy as np
import pathlib
import glob
```

```python
#assigning directory
directory=pathlib.Path("/content/drive/My Drive/BreaKHis 400X/train")
```

```python
# Just needed in case you'd like to append it to an array
data = []

for filename in os.listdir("."):
    if filename.endswith("png"): 
        # Your code comes here such as 
        print(filename)
        data.append(filename)
```

```python
#count of images in the directory given
image_count=len(list(directory.glob('*/*.png')))
#print(list(directory.glob('/*.jpg')))
image_count
```

```python
#creating dictionary of flower species
Breast_images_dict={
    "benign":list(directory.glob('benign/*.png')),
    "malignant":list(directory.glob('malignant/*.png'))
    
}
```

```python
Breast_images_dict
```

```python
#Different species list
keys=["benign","malignant"]
```

```python
#resizing and creating labels using computer vision
resized,labels=[],[]
for Breast_name,images in Breast_images_dict.items():
    for image in images:
        print("hi")
        img=cv2.imread(str(image))
        resized_image=cv2.resize(img,(224,224))
        resized.append(resized_image)
        labels.append(keys.index(Breast_name))
Breast_images_dict
```

```python
#function to print images
def print_image(i,j):
    plt.imshow(i)
    plt.title(keys[j])
```

```python
#splitting data into train and test
#if any integer is given to random_state it will generate and give same set of data each time we run the code and if we give 
#None different sequences will be generated
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(resized,labels,test_size=0.30,random_state=0)
```

```python
print_image(x_train[15],y_train[15])
```

```python
print_image(x_test[6],y_test[6])
```

```python
#normalizing data
x_train_scaled=np.array(x_train)/255
x_test_scaled=np.array(x_test)/255
```

```python
x_train_scaled.shape
```

```python
y_train=np.array(y_train)
y_train.shape
```

```python
model=models.Sequential([
    layers.Conv2D(16,3,padding='same',activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(128,activation="relu"),
    layers.Dense(64,activation="relu"),
    layers.Dense(2,activation="softmax")])
model.compile(
optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(x_train_scaled,np.array(y_train),epochs=50)
```

```python
model.evaluate(x_test_scaled,np.array(y_test))
```

```python
model1=models.Sequential([
    layers.experimental.preprocessing.RandomZoom(0.3),
    layers.Conv2D(16,3,padding='same',activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(128,activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(64,activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(2,activation="softmax"),
    #layers.Dense(5,kernel_regularizer=l1(0.01),activation="softmax")
])
model1.compile(
optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
model1.fit(x_train_scaled,np.array(y_train),epochs=50)
```

```python
model1.evaluate(x_test_scaled,np.array(y_test))
```
